```python
import yt_dlp
import pandas as pd

def get_playlist_metadata(playlist_url):
    ydl_opts = {
        "quiet": True,
        "extract_flat": False,  # Fetch full metadata
        "force_generic_extractor": False,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(playlist_url, download=False)

    videos = []
    if "entries" in info:
        for video in info["entries"]:
            if video:  # Ensure the entry exists
                videos.append(video)  # Save all metadata for each video

    # Convert list of dictionaries to DataFrame
    df = pd.DataFrame(videos)
    df.to_csv("youtube_playlist_full_metadata.csv", index=False)
    print("Full metadata saved to youtube_playlist_full_metadata.csv")
    return df

# Provide the playlist URL
playlist_url = "https://www.youtube.com/playlist?list=PL2hoGhz2jBSrEk4tA0nS72lZjifPPCd8r"
df = get_playlist_metadata(playlist_url)
df
```

```python
import pandas as pd

def format_duration(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    sec = seconds % 60
    return f"{int(hours):02}:{int(minutes):02}:{int(sec):02}"
df_sorted_duration = df.sort_values(by="duration", ascending=False).copy()
df_sorted_duration["formatted_duration"] = df_sorted_duration["duration"].apply(format_duration)
df_sorted_duration[["title", "formatted_duration"]].to_csv("sorted_videos_by_duration.csv", index=False)
df_sorted_duration[["title", "formatted_duration"]].head(10)
```
